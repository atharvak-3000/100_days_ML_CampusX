```python
#creating final dataframe
final = pd.DataFrame()

for j in range(1,11):

    url = (f'https://www.ambitionbox.com/list-of-companies?page={j}')
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36',"Accept": 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7', 'sec-ch-ua': 'Chromium";v="130", "Google Chrome";v="130", "Not?A_Brand";v="99', 'sec-ch-ua-mobile':'?0', 'sec-ch-ua-platform': 'Windows', 'sec-fetch-dest': 'document', 'sec-fetch-mode': 'navigate', 'sec-fetch-site': 'same-origin', 'sec-fetch-user': '?1', 'upgrade-insecure-requests': '1'}
    webpage = requests.get(url,headers=headers).text
    
    soup = BeautifulSoup(webpage,'lxml')

    company=soup.find_all('div',class_='companyCardWrapper')
    name=[]
    rating=[]
    reviews = []
    salaries =[]
    interviews = []
    jobs = []
    benefits = []
    photos = []
    
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div',class_ = 'rating_text').text.strip())
        reviews.append(i.find('span',class_ ='companyCardWrapper__companyRatingCount').text.strip())
        salaries.append(i.find_all('a','companyCardWrapper__ActionWrapper')[1].text.strip())
        interviews.append(i.find_all('a','companyCardWrapper__ActionWrapper')[2].text.strip())
        jobs.append(i.find_all('a','companyCardWrapper__ActionWrapper')[3].text.strip())
        benefits.append(i.find_all('a','companyCardWrapper__ActionWrapper')[4].text.strip())
        photos.append(i.find_all('a','companyCardWrapper__ActionWrapper')[5].text.strip())
    #use find_all when we know it will give multiple answers
    #but her div with class rating text is unique
    
    d = {'name' : name ,'rating':rating , 'reviews':reviews , 'salaries' : salaries , 'interviews' : interviews , 'jobs' : jobs , 'benefits' : benefits ,'photos':photos}
    df = pd.DataFrame(d)
    final = pd.concat([final,df],ignore_index=True)
```
