```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.sample(5)
```

# 1. Categorical Data

```python
# a. countplot
#how much survived
# sns.countplot(x="Survived",data=df)
# df['Survived'].value_counts().plot(kind='bar')
#both are same graphs
sns.countplot(df['Sex'])
```

```python
# b. piechart
#for percentage
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2 .Numerical Data

```python
#a. Histogram
#divide data into bins and then plot them
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins = 5)
```

```python
#b. Distplot
sns.distplot(df['Age'])
# eg- probability of a passenger having age 40 = density * 100 %
#also tells about skewness
```

```python
#c. Boxplot
# gives a 5 number summary
sns.boxplot(df['Fare'])
```

```python
#min and max
df['Age'].min()
```

```python
df['Age'].max()
```

```python
df['Age'].skew()
# 0-->perfectly symmetrical
# positive-->towards right side
# negative-->towards left side
```
